```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

model = tf.keras.models.load_model("best_model2.keras")
model.evaluate(x_test,y_test)
```

```python
class_names = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
        
j = []
for i in range(10) :
    j.append(np.random.randint(0, 10001))

for i in range(10) :
    prediction = model.predict(x_test[j[i]:j[i]+1])
    k = max(prediction[0])
    txt = "prediction : \"" + class_names[prediction[0].tolist().index(k)] + "\" et le numro de l'image : " + str(j[i])
    plt.figure(figsize=(2,2)).text(.5, .05, txt, ha = "center")
    plt.imshow(x_test[j[i]])
    plt.title(class_names[y_test[j[i]][0]])
    plt.axis("off")
```
